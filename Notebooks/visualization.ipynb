```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

file_path = r"C:\\Personal\\coding\S2\sem2\\comvis_hadi\\Experiments\\Experiment1\\FinalTraining\\ranking.csv"
df = pd.read_csv(file_path)

df = df.sort_values(by="mean_f1", ascending=False)

x = np.arange(len(df["model"]))
width = 0.35

plt.figure()
plt.bar(x - width/2, df["mean_accuracy"], width, label="Mean Accuracy")
plt.bar(x + width/2, df["mean_f1"], width, label="Mean F1")

plt.xticks(x, df["model"])
plt.xlabel("Model")
plt.ylabel("Score")
plt.title("Perbandingan Mean Accuracy dan Mean F1 per Model")
plt.ylim(0, 1)
plt.legend()

plt.tight_layout()
plt.show()
```
